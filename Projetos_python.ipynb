# API - E um lugar para disponilizar recursos e/ou funcionalidades

# 1. Objetivo - Criar um api que disponibiliza a consulta, criação, edição e exclusao de livros

# 2. URL base - Localhost

# 3. EndPoints -

# - localhost/livros/id (Delete)

```python
## 1. Instalação das versões atualizadas
!pip install flask pyngrok --quiet

from flask import Flask, jsonify
from pyngrok import ngrok
import getpass

# 2. Configuração do Authtoken (Cole o token que você pegou no site)
print("Cole seu Authtoken do Ngrok abaixo:")
conf_token = getpass.getpass()
ngrok.set_auth_token(conf_token)

app = Flask(__name__)

livros = [
    {'id': 1, 'título': 'Percy Jackson e o Mar de Monstros', 'autor': 'Rick Riordan'},
    {'id': 2, 'título': 'Percy Jackson e o Ladrão de Raios', 'autor': 'Rick Riordan'},
    {'id': 3, 'título': 'Percy Jackson e a Maldição do Titã', 'autor': 'Rick Riordan'},
]
#consultar todos
@app.route('/livros', methods=['GET'])
def obter_livros():
    return jsonify(livros)

#consultar(ID)
@app.route('/livros/<int:id>', methods=['GET'])
def obter_livro_por_id(id):
    for livro in livros:
        # Se o id do livro atual for igual ao id passado na URL
        if livro.get('id') == id:
            return jsonify(livro)

# 3. Criando o túnel manualmente (o jeito que não dá erro)
try:
    # Abre o túnel na porta 5000
    public_url = ngrok.connect(5000).public_url
    print(f" * API ATIVA!")
    print(f" * Clique aqui para acessar: {public_url}/livros")

    # Roda o Flask
    app.run(port=5000)
finally:
    # Garante que o túnel fecha se você parar o código
    ngrok.kill()
```
